# Transfermarkt 크롤링

- 첫 페이지 25명 정보

```python
# 필요한 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# requests.get()으로 url 정보 요청하기
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}

url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop"

r = requests.get(url, headers = headers)

r.status_code
```

```python
# BeautifulSoup()으로 웹페이지 분석 준비하기

soup = BeautifulSoup(r.content, 'html.parser')    # r.content 대신 r.text도 가능
print(soup)
```

```python
# 선수들의 정보가 담긴 태그와 클래스 찾기

player_info = soup.find_all('tr', class_ = ['odd', 'even'])
```

```python
# 첫 번째 요소 확인하기
print(player_info[0])

# 전체 개수 확인하기
print(len(player_info))
```

```python
# 7개 정보를 담을 빈 리스트 만들기 number, name, position, age, nation, team, value

number = []
name = []
position = []
age = []
nation = []
team = []
value = []
```

```python
# player_info에서 'td' 태그만 모두 찾기

for info in player_info:
    player_td = info.find_all('td')
# 해당 정보를 찾아서 각 리스트에 .append로 추가하기
    number.append(player_td[0].text)
    name.append(player_td[3].text)
    position.append(player_td[4].text)
    age.append(player_td[5].text)
    nation.append(player_td[6].img['alt'])
    team.append(player_td[7].a['title'])
    value.append(player_td[8].text.strip())    # strip() -> 앞뒤 공백 제거
```

```python
# pd.DataFrame()으로 저장하기

df = pd.DataFrame(
    {"number":number,
     "name":name,
     "position":position,
     "age":age,
     "nation":nation,
     "team":team,
     "value":value}
)

df
```

```python
# df.to_csv()를 이용하여 csv 파일로 저장하기
# csv(comma-separated values) : 데이터를 쉼표(,)로 구분한 텍스트 파일

df.to_csv('transfermarkt25.csv', index = False)
```
